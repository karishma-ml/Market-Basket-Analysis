```python
df["Quantity"]=1 # initialize by 1
```

```python
df.groupby(["Member_number","itemDescription"])["Quantity"].sum()
```

```python
df.groupby(['Member_number' , 'itemDescription'])['Quantity'].sum().unstack()# convert into columns 
```

```python
transaction=df.groupby(['Member_number' , 'itemDescription'])['Quantity'].sum().unstack().reset_index().set_index('Member_number')
```

```python
transaction
```

```python
transaction = transaction.fillna(0)
```

```python
transaction
```

```python
def encode (df):
    return 1 if df > 0 else 0

basket = transaction.applymap(encode)
```

```python
basket
```

```python
# machine leraning extension
from mlxtend.frequent_patterns import apriori
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemset = apriori(basket,min_support=0.06,use_colnames=True)
```

```python
rules = association_rules(frequent_itemset , metric = 'lift' , min_threshold = 1)
```

```python
rules.head()
```

```python
rules[(rules['confidence']>0.4) & (rules['lift']>1)]
```
